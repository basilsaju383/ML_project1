```python
import pandas as pd
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

# Create a dataframe with 4 features
data = pd.DataFrame(np.random.multivariate_normal(mean=[10, 10, 10, 10], cov=[(5, 1.5, 1, 1), (1.5, 5, 1, 1), (1, 1, 5, 1.5), (1, 1, 1.5, 5)], size=100), columns=['f1', 'f2', 'f3', 'f4'])

# Add anomalies
data.loc[50, ['f1', 'f2', 'f3', 'f4']] = [25, 25, 25, 25]
data.loc[51, ['f1', 'f2', 'f3', 'f4']] = [15, 15, 15, 15]
data.loc[52, ['f1', 'f2', 'f3', 'f4']] = [18, 18, 18, 18]
```

```python
# Apply Local Outlier Factor
clf = LocalOutlierFactor(n_neighbors=20, contamination='auto')
scores_lof = clf.fit_predict(data)
```

```python
# Visualize the results
fig, ax = plt.subplots(figsize=(8, 6))

data.plot(kind='scatter', x='f1', y='f2', c=scores_lof, cmap='coolwarm', ax=ax)
ax.set_title('Local Outlier Factor - Anomaly Detection')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
plt.show()
```
